### Polygon Tomfoolery, Count metro stations per neighbourhood

```python
def check_point(lat,long,poly):
    '''
    Check if GPS coordinates are inside of a polygon 
    
    '''
    point = Feature(geometry = Point((long, lat)))
    polygon = Polygon([list(poly.exterior.coords)])
         
    return boolean_point_in_polygon(point, polygon)
```

```python
station_hood = []

#Find the neighborhood specific metro stations belong to
for station in metro_coords.index:
    station_latitude = metro_coords['latitude'][station]
    station_longitude = metro_coords['longitude'][station]
    for neighborhood in neighbor_polygons.index:
        neighbor_poly = neighbor_polygons[neighborhood]
        if check_point(station_latitude,station_longitude,neighbor_poly):
            station_hood.append((neighborhood,station))
```

```python
# Check if the neighborhood assignment make sense, looks it does!
station_hood[0:15]
```

```python
# Count the number of stations per neighborhood. 
neighborhood_numstations = []
for key, group in groupby(station_hood, lambda x: x[0]):
    neighborhood_numstations.append((key,len(list(group))))
    
neighborhood_numstations[0:10]
```

```python
# Turning the list above into a pandas seris
stations_dict = {k:v for k,v in neighborhood_numstations}

stations_dict
```

```python
# Add the number of venue rating  column 

# create an array that maps the neighborhood to the number of venues it has 
stations_arr = np.array([stations_dict[hood]\
                                 if hood in stations_dict.keys() else 0 for hood in df_main['neighborhood'].to_list()])

df_main['num_stations'] = stations_arr
df_main.head()
```
